__1)__ Faça um programa que lê uma string do teclado e a exibe invertida na tela.

__2)__ Faça um programa que pede para o usuário digitar 2 palavras. O seu programa deverá responder se elas são palíndromos uma da outra, ou seja, se ao inverter uma delas, temos a outra. 

Exemplo: as palavras "roma" e "amor" são palíndromos.

```python
palavra1 = input('Digite a primeira palavra: ')
palavra2 = input('Digite a segunda palavra: ')

string_normal = palavra1
palavra_temp = []
for i in range(len(string_normal), 0,-1):
  palavra_temp.append(string_normal[i-1])
palavra_invertida =  ''.join(palavra_temp)

if palavra_invertida == palavra2:
  print('As duas palavras são palíndromos, pois invertidas elas são iguais.')
  print(f'Palavra1: {palavra1}    =>    Invertida: {palavra_invertida}    =>    Palavra2: {palavra2}.')
else:
  print('As duas palavras NÃO são palíndromos, pois invertidas elas não são iguais.')
  print(f'Palavra1: {palavra1}    =>    Invertida: {palavra_invertida}    =>    Palavra2: {palavra2}.')
```

__3)__ Faça um programa que lê duas palavras do teclado e responde se elas são **anagramas** uma da outra, isto é, se podemos escrever uma delas apenas mudando a ordem das letras da outra.

Exemplo: "ator" e "rota", "rota" e "tora", "lama" e "mala". 

```python
def remove_letra(partes_palavra, letra):
  palavra = ''.join(partes_palavra)
  palavra.replace(letra, '')
  return palavra.split()

def extrair_letras(palavra):
  partes_palavra = []
  for i in range(len(palavra)):
    partes_palavra.append(palavra[i])
  return partes_palavra

def main():
  palavra1 = input('Digite a primeira palavra: ')
  palavra2 = input('Digite a segunda palavra: ')

  partes_palavra1 = extrair_letras(palavra1)
  partes_palavra2 = extrair_letras(palavra2)

  resultado_anagrama = False

  if len(palavra1) >= len(palavra2):
    resultado_anagrama = analisar_palavras(partes_palavra1, partes_palavra2)
  else:
    resultado_anagrama = analisar_palavras(partes_palavra2, partes_palavra1)

  if resultado_anagrama and len(palavra1) >= len(palavra2):
    print('A palavra 2 é anagrama da palavra 1.')
  elif resultado_anagrama and len(palavra2) >= len(palavra1):
    print('A palavra 1 é anagrama da palavra 2.')
  else:
    print('As palavras NÂO são anagramas uma da outra.')
  return

def analisar_palavras(letras1, letras2):
  validacao = False
  for i in range(len(letras1)):
    letra_palavra1 = 0
    letra_palavra2 = 0

    # Verifica a quantidade de vezes que a letra se repeta nas duas palavras.
    quant_letra_palavra1 = letras1.count(letras1[i])
    quant_letra_palavra2 = letras2.count(letras1[i])

    # Remove a letra da palavra, pois já foi analisada
    remove_letra(letras1, letras1[i])
    remove_letra(letras2, letras1[i])

    if quant_letra_palavra1 == quant_letra_palavra2:
      validacao = True
    else:
      continue
  return validacao

main()
```

__4)__ **Cifra de César**

A cifra de César é uma forma rudimentar de criptografia onde todas as letras são deslocadas uma quantidade pré-determinada de vezes no alfabeto. Por exemplo, a frase "Abra a porta" com 3 deslocamentos vira "Deud d sruwd". 

Caso os deslocamentos estourem o tamanho do alfabeto, eles "dão a volta": no caso anterior, com 3 deslocamentos, se alguma palavra tivesse a letra "z", ela se tornaria "c".

Faça um programa que pergunta para o usuário uma frase para ser criptografada e a quantidade de deslocamentos desejada. Seu programa irá mostrar a mensagem cifrada na tela.



```python
def extrair_letras(frase):
  partes_frase = []
  for i in range(len(frase)):
    partes_frase.append(frase[i])
  return partes_frase

def criptografia(frase, deslocamento):
  alfabeto = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
  frase_temp = []
  for i in range(len(frase)):
    if (frase[i] not in alfabeto) or (frase[i] == ''):
      frase_temp.append(frase[i])
    else:
      index_letra_original = alfabeto.index(frase[i]) # pega o index da letra original referente ao alfabeto
      index_letra_cifrada = index_letra_original + deslocamento # calcula o index da letra cifrada, deslocada de acordo com o solicitado
      if index_letra_cifrada > 25:
        index_letra_cifrada = index_letra_cifrada - 25 -1 # o -1 é para corrigir o index das listas que sempre começam em zero
      frase_temp.append(alfabeto[index_letra_cifrada])
  return frase_temp

def main():
  frase = input('Digite a frase que desaja criptografar (somente letras): ').lower()
  partes_frase = extrair_letras(frase)
  qntd_deslocamento = int(input('Digite qual a quantidade de deslocamentos desejatos (apenas números inteiros): '))
  frase_temp = criptografia(partes_frase, qntd_deslocamento)
  frase_criptografada = ''.join(frase_temp)
  return frase_criptografada
  
main()
```

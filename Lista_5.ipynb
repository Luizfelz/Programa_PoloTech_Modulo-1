__1)__ Faça um programa que receba um dicionario simples de um número associado a uma chave e retorne o diciona inverso (onde era chave virá valor e vice-versa).

Exemplos:

````python
dict1 = {1:3,2:5,3:7,4:0}
values = []
keys = []
keys = list(dict1.keys())
values = list(dict1.values())
dict_reverso = {}
for i in range(len(keys)):
  dict_reverso[values[i]] = [keys[i]]

dic2 = {1: 2, 2: 0, 3: 0, 4: 2}
keys2 = sorted(list(dic2.keys()))
values2 = sorted(list(dic2.values()))
newDic = {}
for i in range(len(dic2)):
  if newDic.get(values2[i]):
    newDic[values2[i]] += [keys2[i]]   
  else:
    newDic[values2[i]] = [keys2[i]]

print(f'Dicionário 1: {dict1}.\tDicionário 1 reverso: {dict_reverso}.')
print(f'Dicionário 2: {dic2}.\tDicionário 2 reverso: {newDic}.')
````

__2)__ Desenvolva uma função que retorne uma lista com todas as chaves de um dicionário que contém o mesmo valor. Por exemplo no dicionário a seguir:

<br>

```python
{'a': 3, 'b': 1, 'c': 2, 'd': 3}
```

<br>

A função deve retorna os valores ['a', 'd'] caso o valor procurado seja o 2. E caso não tenha o valor procurado deve-se retornar uma lista vazia.

````python
dict3 = {'a': 3, 'b': 1, 'c': 2, 'd': 3}

def retorna_chaves_de_valor(dicionario):
  chaves3 = list(dicionario.keys())
  valores3 = list(dicionario.values())
  newDic = {}
  for i in range(len(dicionario)):
    if newDic.get(valores3[i]):
      newDic[valores3[i]] += [chaves3[i]]   
    else:
      newDic[valores3[i]] = [chaves3[i]]

  print('Os valores e suas respectivas chaves são: ')
  print('Valor\tChaves')
  for i in newDic.keys():
      print(f'  {i}\t{newDic[i]}')

retorna_chaves_de_valor(dict3)
````

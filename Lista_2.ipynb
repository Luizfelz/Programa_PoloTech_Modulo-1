__1)__ Desenvolva um código que recebe o valor de peso (em kgs) e a altura (em metros) de uma pessoa e devolva um print do IMC e em qual categoria de IMC ela se encontra. Para calcular deve-se utilizar a fórmula a seguir:

<br>

$$ IMC = \frac{peso}{(altura)^2} $$

<br>

E as faixas de IMC são classificadas segundo a tabela abaixo:

<br>

|      **IMC**      |      **Faixa**     |
|:-----------------:|:------------------:|
|       < 18,5      |       Subpeso      |
| entre 18,6 e 24,9 |     Peso Ideal     |
| entre 25,0 e 29,9 |    Acima do Peso   |
| entre 30,0 e 34,9 |  Obesidade grau I  |
| entre 35,0 e 39,9 |  Obesidade grau II |
|      >= 40,0      | Obesidade grau III |

```python
# Exemplos
# Teste 1 = peso = 130 kg e altura = 1,72 m 
# Resultado : IMC = 43,9, Obesidade grau III

# teste II = peso = 79 kg e altura = 1,69 m
# resultado: IMC = 27,6, Acima do Peso

# Não precisa se preocupar em arredondar os valores, pode trazer com todas as casas!
```

```python
massa = float(input('Digite sua massa (em kg): '))
altura = float(input('Digite sua altura (em metros): '))
imc = massa / (altura ** 2)

if imc < 18.5:
  print(f'O IMC é = {imc} e a situação é: Subpeso.')
elif ((imc >= 18.5) and (imc < 25)):
  print(f'O IMC é = {imc} e a situação é: Peso ideal.')
elif ((imc >= 25) and (imc < 30)):
  print(f'O IMC é = {imc} e a situação é: Acima do peso.')
elif ((imc >= 30) and (imc < 35)):
  print(f'O IMC é = {imc} e a situação é: Obesidade grau I.')
elif ((imc >= 35) and (imc < 40)):
  print(f'O IMC é = {imc} e a situação é: Obesidade grau II.')
else:
  print(f'O IMC é = {imc} e a situação é: Obesidade grau III')


'''
  Caso necessário, o "notebook" com todos os 5 exercícios está em anexo nesta questão 1.
'''
```

__3 ao 5)__ O colégio Schmalkalden tem um dos sistemas mais rigorosos e complexos para a avaliação dos alunos, onde até os alunos tem dificuldades para entender realmente se de fato foram aprovados ou ficaram devendo alguma coisa. Para isso, você deverá desenvolver um código que auxilie os alunos a saberem se foram aprovados ou não. O programa deverá receber como input dos alunos as seguintes informações:

<br>

- Se o aluno é bolsista ou não?
- Qual a nota das 4 provas bimestrais que o aluno tirou? (Aqui serão 4 inputs de notas que variam entre 0 e 10, podendo ser decimal);
- Quantas listas de exercícios oram entregues ao longo do ano? (de um total de 20 listas no ano)
- Qual foi a nota no trabalho anual desenvolvido pelo aluno?
- Quantas faltas o aluno teve ao longo do ano?

<br>

Para calcular a média dos alunos, deve-se seguir a mesma regra do exercício da aula passada, conforme a equação abaixo:

<br>

$$ \bar{x} = \frac{w1*P + w2*L + w3*T}{w1 + w2 + w3}$$

<br>

Onde P é a média aritmética das 4 provas que o aluno fez, L vai ser a nota em relação as entregas das listas de exercícios (cada lista entregue vale 0,5 ponto) e T é a nota do trabalho. Agora, os critérios de avaliação dos alunos são conforme descrito abaixo:

- __Alunos Bolsistas__: 
- Os pesos serão respectivamente $w1 = 7$, $w2 = 6$ e $w3 = 9$
- Deve ter menos do que 5 faltas para ser aprovado, se tiver entre 5 e 10 faltas o aluno deve entregar um trabalho complementar, e por fim mais de 10 faltas reprovado diretamente
- A média das provas tem que ser maior do 8.5 para ser aprovado, entre 6.5 e 8.5 deve fazer uma prova de recuperação e menor que 6.5, reprovado por prova;
- As listas devem ter mais de 75% entregue para aprovado em listas, entre 10 e 15 listas o aluno deve entregar uma lista complementar, e se tiver menos que 10 listas entregues reprovado por lista;
- A nota do trabalho deve ser maior que 6 para ser aprovado, caso contrário será reprovado por trabalho;
****************************************************************************
- __Alunos Não Bolsistas__: 
- Os pesos serão respectivamente $w1 = 3$, $w2 = 8$ e $w3 = 2$
- Deve ter menos do que 10 faltas para ser aprovado, se tiver entre 10 e 15 faltas o aluno deve entregar um trabalho complementar, e por fim mais de 15 faltas reprovado diretamente
- A média das provas tem que ser maior do 6.5 para ser aprovado, entre 4.5 e 6.5 deve fazer uma prova de recuperação e menor que 4.5, reprovado por prova;
- As listas devem ter mais de 50% entregue para aprovado em listas, entre 5 e 10 listas o aluno deve entregar uma lista complementar, e se tiver menos que 5 listas entregues reprovado por lista;
- A nota do trabalho deve ser maior que 6 para ser aprovado, caso contrário será reprovado por trabalho;

Então depois de todos os inputs dado pelo o aluno, o seu programa deve retornar:
- Média das provas
- Status da Prova (Aprovado, Prova de Recuperação ou Reprovado por Prova)
- Status da Presença (Aprovado em Presença, Trabalho COmplementar ou Reprovado por Presença)
- Nota das Listas
- Status das Listas ( Aprovado, Entrega de Lista COmplementar ou Reprovado Por Lista)
- Status do Trabalho (aprovado ou reprovado)
- Média Final do aluno

```python
bolsista = input('O aluno é bolsista [s/n]? ')
nota1 = float(input('Digite a nota do bimestre 1: '))
nota2 = float(input('Digite a nota do bimestre 2: '))
nota3 = float(input('Digite a nota do bimestre 3: '))
nota4 = float(input('Digite a nota do bimestre 4: '))
qntd_listas = int(input('Quantas listas foram entregues (de 0 a 20)? '))
trab_anual = float(input('Qual foi a nota do trabalho anual? '))
faltas = int(input('Quantas faltas teve o aluno? '))

media_arit = (nota1 + nota2 + nota3 + nota4) / 4
print(f'A média das provas foi = {media_arit}.')

if bolsista == 's':
  w1 = 7
  w2 = 6
  w3 = 9
  
  if media_arit > 8.5:
    print('Aluno aprovado pela média das provas.')
  else:
    if media_arit >= 6.5 and media_arit <= 8.5:
      print('O aluno deve fazer uma prova de recuperação.')
    else:
      print('O aluno reprovou por média.')
  
  if faltas > 10:
    print(f'Faltas = {faltas}. O aluno foi reprovado por faltas')
  else:  
    if faltas >=5 and faltas <= 10:
      print(f'Faltas = {faltas}. O aluno deve entregar um trabalho complementar para justificar faltas.')
    else:
      print(f'Faltas = {faltas}. O aluno foi aprovado por faltas.')
  
  print(f'A nota das listas é = {qntd_listas * 0.5}.')
  if qntd_listas > 15:
    print('Aluno aprovado pelas listas.')
  else:
    if qntd_listas >= 10 and qntd_listas <= 15:
      print('Aluno deve entregar uma lista complementar.')
    else:
      print('Aluno reprovado pelas listas.')
  
  if trab_anual > 6:
    print('Aluno aprovado no trabalho anual.')
  else:
    print('Aluno reprovado no trabalho anual.')
  
  media_final = ((w1 * media_arit) + (w2 * qntd_listas * 0.5) + (w3 * trab_anual)) / (w1 + w2 + w3)
  print(f'A média final do aluno é = {media_final}.')

# -------------------------------------------------------------------------------------------------------------
else:
  if bolsista == 'n':
    w1 = 3
    w2 = 8
    w3 = 2
    
    if media_arit > 6.5:
      print('Aluno aprovado pela média das provas.')
    else:
      if media_arit >= 4.5 and media_arit <= 6.5:
        print('O aluno deve fazer uma prova de recuperação.')
      else:
        print('O aluno reprovou por média.')
    
    if faltas > 15:
      print(f'Faltas = {faltas}. O aluno foi reprovado por faltas')
    else:
      if faltas >=10 and faltas <= 15:
        print(f'Faltas = {faltas}. O aluno deve entregar um trabalho complementar para justificar faltas.')
      else:
        print(f'Faltas = {faltas}. O aluno foi aprovado por faltas.')
    
    print(f'A nota das listas é = {qntd_listas * 0.5}.')
    if qntd_listas > 10:
      print('Aluno aprovado pelas listas.')
    else:
      if qntd_listas >= 5 and qntd_listas <= 10:
        print('Aluno deve entregar uma lista complementar.')
      else:
        print('Aluno reprovado pelas listas.')
    
    if trab_anual > 6:
      print('Aluno aprovado no trabalho anual.')
    else:
      print('Aluno reprovado no trabalho anual.')
    
    media_final = ((w1 * media_arit) + (w2 * qntd_listas * 0.5) + (w3 * trab_anual)) / (w1 + w2 + w3)
    print(f'A média final do aluno é = {media_final}.')

  else:
    print('Opção de bolsista digitada é inválida!')
```

```python
'''
  O exercício feito utilizando apenas ifs torna mais complexo o raciocínio, uma vez que é necessário que o programa seja feito de tal forma que evite que mais de um resultado
  seja válido par auma mesma condição. Ou seja, caso o aluno (por exemplo) tire nota 9, é necessário que seja o programa seja feito de tal forma que avalie cada condição levando 
  em conta o menor valor possível como resposta para a nota (neste caso, a reprovação, caso a nota seja menor que 4.5), na sequência a necessidade de prova de recuperação (caso a
  nota ficasse entre 4.5 e 6.5) e, por fim, para a nota maior que 6.5 (aprovação). Dessa forma, apenas a última condição seria "acionada" pelo programa, uma vez que 9 é maior ou
  igual a 6.5, mas não satisfaz as condições anteriores. Portanto, é possível sim que seja feito o programa utilizando-se apenas ifs, mas isso faz com que o programa fique
  extremamente susceptível a erros de lógica. 

  O programa, utilizando-se apenas ifs, é demonstrado abaixo. É possível perceber, se comparado aos programas dos exercícios 3 e 4, que as condições foram invertidas, para
  satisfazer a lógica sequencial necessária para que o programa funcione corretamente (neste exercício específico).
'''

bolsista = input('O aluno é bolsista [s/n]? ')
nota1 = float(input('Digite a nota do bimestre 1: '))
nota2 = float(input('Digite a nota do bimestre 2: '))
nota3 = float(input('Digite a nota do bimestre 3: '))
nota4 = float(input('Digite a nota do bimestre 4: '))
qntd_listas = int(input('Quantas listas foram entregues (de 0 a 20)? '))
trab_anual = float(input('Qual foi a nota do trabalho anual? '))
faltas = int(input('Quantas faltas teve o aluno? '))

media_arit = (nota1 + nota2 + nota3 + nota4) / 4
print(f'A média das provas foi = {media_arit}.')

if bolsista == 's':
  w1 = 7
  w2 = 6
  w3 = 9
  
  if media_arit < 6.5:
    print('Aluno reprovado pela média das provas.')
  if media_arit >= 6.5 and media_arit <= 8.5:
    print('O aluno deve fazer uma prova de recuperação.')
  if media_arit > 8.5:
    print('O aluno reprovou por média.')
  
  if faltas < 5:
    print(f'Faltas = {faltas}. O aluno foi aprovado por faltas')
  if faltas >=5 and faltas <= 10:
    print(f'Faltas = {faltas}. O aluno deve entregar um trabalho complementar para justificar faltas.')
  if faltas > 10:
    print(f'Faltas = {faltas}. O aluno foi reprovado por faltas.')
  
  print(f'A nota das listas é = {qntd_listas * 0.5}.')
  if qntd_listas < 10:
    print('Aluno reprovado pelas listas.')
  if qntd_listas >= 10 and qntd_listas <= 15:
    print('Aluno deve entregar uma lista complementar.')
  if qntd_listas > 15:
    print('Aluno aprovado pelas listas.')
  
  if trab_anual <= 6:
    print('Aluno reprovado no trabalho anual.')
  if trab_anual > 6:
    print('Aluno aprovado no trabalho anual.')
  
  media_final = ((w1 * media_arit) + (w2 * qntd_listas * 0.5) + (w3 * trab_anual)) / (w1 + w2 + w3)
  print(f'A média final do aluno é = {media_final}.')

# -------------------------------------------------------------------------------------------------------------

if bolsista == 'n':
  w1 = 3
  w2 = 8
  w3 = 2
    
  if media_arit < 4.5:
    print('Aluno reprovado pela média das provas.')
  if media_arit >= 4.5 and media_arit <= 6.5:
    print('O aluno deve fazer uma prova de recuperação.')
  if media_arit > 6.5:
    print('O aluno aprovado por média.')
    
  if faltas < 10:
    print(f'Faltas = {faltas}. O aluno foi aprovado por faltas')
  if faltas >=10 and faltas <= 15:
    print(f'Faltas = {faltas}. O aluno deve entregar um trabalho complementar para justificar faltas.')
  if faltas > 15:
    print(f'Faltas = {faltas}. O aluno foi reprovado por faltas.')
    
  print(f'A nota das listas é = {qntd_listas * 0.5}.')
  if qntd_listas < 5:
    print('Aluno reprovado pelas listas.')
  if qntd_listas >= 5 and qntd_listas <= 10:
    print('Aluno deve entregar uma lista complementar.')
  if qntd_listas > 10:
    print('Aluno aprovado pelas listas.')
    
  if trab_anual <= 6:
    print('Aluno reprovado no trabalho anual.')
  if trab_anual > 6:
    print('Aluno aprovado no trabalho anual.')
    
  media_final = ((w1 * media_arit) + (w2 * qntd_listas * 0.5) + (w3 * trab_anual)) / (w1 + w2 + w3)
  print(f'A média final do aluno é = {media_final}.')

if bolsista != 's' and bolsista != 'n':
  print('Opção de bolsista digitada é inválida!')
```
